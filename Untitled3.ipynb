```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
codon_variant = pd.read_csv('results/variants/codon_variants.csv', usecols=['barcode','library','barcode','codon_substitutions','aa_substitutions','n_codon_substitutions','n_aa_substitutions'])
filtered_df = codon_variant.loc[(codon_variant['library'] == 'LibB') & (codon_variant['codon_substitutions'] != 'neut_standard')]
```

```python
filtered_df


```

```python
# Drop duplicate rows based on 'column_name'
unique_rows = filtered_df.drop_duplicates(subset='aa_substitutions')
unique_rows
```

```python
# Extract the 'aa_substitutions' column and split the mutations
mutations = filtered_df['aa_substitutions'].dropna().str.split()

# Flatten the list of lists into a single list
flattened_mutations = [mutation for sublist in mutations for mutation in sublist]

# Count unique mutations
unique_mutations = pd.Series(flattened_mutations).unique()

# Print the number of unique single mutations
print(f"Number of unique single mutations: {len(unique_mutations)}")
print("Unique single mutations:", unique_mutations)
```

```python
max_n_aa_substitutions = filtered_df['n_aa_substitutions'].max()
max_n_aa_substitutions 
```

```python
# Plot overlapping histograms
plt.figure(figsize=(8, 4.8))
sns.histplot(filtered_df['n_aa_substitutions'], bins=range(0, 11), kde=False, color='#008080', label='n_aa_substitutions',legend=False, alpha=0.6)
# sns.histplot(filtered_df['n_codon_substitutions'], bins=range(0, 11), kde=False, color='#FF6347', label='n_codon_substitutions', alpha=0.6)

plt.title('')
plt.xlabel('')
plt.ylabel('')
plt.xlim(0, 10)  # Limit x-axis to 10 bins
plt.savefig('library_n_aa_dist.png', format='png', dpi=300,bbox_inches='tight')
plt.show()
```
